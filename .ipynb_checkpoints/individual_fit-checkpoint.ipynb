```python
#this function prints out the output of a fit in the same way that autofit.py does,
#so you can copy and paste the output from here into your autofit.py output file
def fileprinter(ft):
    ymod, p1, p1cov, yieldarr, yerrarr = ft[0], ft[1], ft[2], ft[3], ft[4]
    off, sig, r, beta = p1[-4], p1[-3], p1[-2], p1[-1]
    nopeaks = int((len(p1) - 4)/2)
    swidth, sr, sbeta = np.sqrt(p1cov[-3][-3]), np.sqrt(p1cov[-2][-2]), np.sqrt(p1cov[-1][-1])
    
    print("  - " + file[-19:-17] + " degrees " + str(lb) + " - " + str(ub))
    
    for i in range(nopeaks):
        #get the right thing from each array of relevant parameters
        Y = yieldarr[i]
        sY = yerrarr[i]
        pos = p1[2*i + 1]
        spos = np.sqrt(p1cov[2*i + 1][2*i + 1]) #errors come from the square root of the diagonals of the covariance matrix
        print(pos, spos, Y, sY, sig, swidth, r, sr, beta, sbeta)
```

```python
#list the files
dire = "spectra/*"
filelist = glob.glob(dire)
filelist
```

```python
file = filelist[0] #Change this to change the file you want to read in


#plot this spectrum out
x, y = ad.file_reader(file, delete_0 = False)
fig, ax, = plt.subplots(figsize = (15,7), dpi = 200)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)

ax.set_xlabel('Ejectile Momentum (arbitrary units)', fontsize = 26)
ax.set_ylabel('Frequency', fontsize = 26)
ax.tick_params(axis="both", labelsize=26)

#ax.scatter(-x + max(x),y, marker = '.')
#ax.scatter(x,y, marker = '.')
ax.plot(x,y)

#fig.savefig('scatter.png')

#I put this in because at Munich the number of counts in the zero channel gives the dead time correction.

print(y[0])
print(np.sqrt(y[0]))
print(np.sum(y) - y[0])
print(np.sqrt(np.sum(y) - y[0]))
```

```python
x, y = ad.file_reader(file, delete_0 = False)

#select bounds for the fit
lb = 1
ub = 1000

y = y[(x < ub) & (x > lb)]
y
x = x[(x < ub) & (x > lb)]

#plot them
#plt.errorbar(x,y, np.sqrt(y + 1), fmt = '.')
plt.plot(x,y)

#Can get centroid and area of a peak directly if it isn't possible to get this through fitting

#centroid of this region and its uncertainty
print(np.sum(x * y)/np.sum(y))
print(np.sqrt(np.sum(y * (x - np.sum(x * y)/np.sum(y))**2)/np.sum(y))/np.sqrt(np.sum(y)))

#Total number of counts in this region and their uncertainties
print(np.sum(y))
print(np.sqrt(np.sum(y)))
```

```python
#Advisory list of peak positions by doing a continuous wavelet tranform!

#set bounds of widths it's looking for
width_lb = 8
width_ub = 12

peak_positions, region_positions = sig.peak_finder([[x,y]], width_lb , width_ub)
print(peak_positions)
```

```python
#set array of peak positions
muarr = np.array([350.5]) - 10
#muarr = np.array(peak_positions)
width_upperbound = 20
si =0.1

ft = fit.fit(x,y,muarr,si,width_upperbound, rbfix = False,r = 20,beta =20, posfix  = False, background = True)
plt.ylim(0)
#plt.xlim(460,555)
plt.xlabel("Channel Number")
plt.ylabel("Count Number")
```
